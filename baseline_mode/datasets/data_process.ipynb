```python
import pandas as pd

action = pd.read_table('action.csv', sep=',')
item = pd.read_table('item.csv', sep=',')
# action = action.iloc[0:300]
action.sort_values(by=['user_id', 'action_time'], ascending=True, inplace=True)
action.drop([ 'module_id'], axis=1, inplace=True)

a = action.groupby(['user_id']).filter(lambda x: (len(x) > 50))

a.reset_index(inplace=True)
a.drop(['index'], axis=1, inplace=True)

sku_to_cate = {}
for i in range(len(item)):
    sku_to_cate[item['sku_id'][i]] = item['cate'][i]

sku_to_brand = {}
for i in range(len(item)):
    sku_to_brand[item['sku_id'][i]] = item['brand'][i]

sku_to_shop = {}
for i in range(len(item)):
    sku_to_shop[item['sku_id'][i]] = item['shop_id'][i]

for i in range(len(a)):
    if a['sku_id'][i] not in sku_to_cate.keys():
        a['sku_id'][i] = -1

for i in range(len(a)):
    if a['sku_id'][i] not in sku_to_shop.keys():
        a['sku_id'][i] = -1

for i in range(len(a)):
    if a['sku_id'][i] not in sku_to_brand.keys():
        a['sku_id'][i] = -1

a.drop(a[a['sku_id'] == -1].index, inplace=True)
a.reset_index(inplace=True)
a_part = a.iloc[0:1000000]
order = ['user_id','sku_id','type','action_time']
a_part = a_part[order]
a_part.to_csv('../raw_jdata/user_action_part.txt', sep=' ', index=False)
```
